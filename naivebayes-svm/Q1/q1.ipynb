```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
training_set = pd.read_csv('Corona_train.csv')
training_set = training_set.to_numpy().T

training_set_x = training_set[2]
training_set_y = training_set[1]
```

```python
# word - idx, pos_ct, neu_ct, neg_ct
vocab = {}

labels = {"Positive": 1,
         "Neutral":2,
         "Negative":3}

pos, neg, neut = 0, 0, 0
poslen, neglen, neutlen = 0, 0, 0
for i in range(training_set_x.shape[0]):
    msg = training_set_x[i]
    label = training_set_y[i]
    words = msg.split()
    if (label=='Positive'):
        pos += 1
        poslen += len(words)
    elif (label=="Negative"):
        neg += 1
        neglen += len(words)
    else:
        neut += 1
        neutlen += len(words)
    
    for word in words:
        if not (word in vocab.keys()):
            vocab[word] = [len(vocab), 0, 0, 0]
        vocab[word][labels[label]] += 1
```

```python
theta_pos = []
theta_neg = []
theta_neut = []
vocab_len = len(vocab)
for word in vocab:
    positive = (vocab[word][1]+1)/(poslen + vocab_len)
    negative = (vocab[word][3]+1)/(neglen + vocab_len)
    neutral = (vocab[word][2]+1)/(neutlen + vocab_len)
    theta_pos.append(positive)
    theta_neg.append(negative)
    theta_neut.append(neutral)

len(theta_neg), theta_neg, theta_pos, theta_neut
```
